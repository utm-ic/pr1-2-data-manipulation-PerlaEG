#### **Paso 3. Limpieza de datos**

Se identifica el número de datos nulos en cada columna, también se busca  identificar el número de registros duplicados en el dataset

```python
missed_values = df_mh.isnull().sum()    #Número de datos nulos por columna
print(missed_values)
```

```python
total_cells = df_mh.size  #Número de celdas del dataset
total_missing = missed_values.sum()  #Número de celdas con datos nulos
missed_percentage = (total_missing/total_cells) * 100

print ("Porcentaje de datos nulos en el dataframe: {0:.2f}%".format(missed_percentage))
```

```python
duplicated = df_mh.duplicated().sum()   #Número de registros duplicados
```

#### **Paso 4. Análisis Estadístico**

Cálculo de algunas estadísticas básicas para tres columnas: "Anxiety_Score", "Depression_Score" y "Stress_Level"


```python
# Cálculo de la media, mediana y la desviación estándar de las variables
# "Anxiety_Score", "Depression_Score" y "Stress_Level"
#Utilizando el método agg de pandas
subset_df_mh = df_mh[['Anxiety_Score','Depression_Score','Stress_Level']]
totales=subset_df_mh.agg(['mean','median','std'])
print(totales)
```

#### **Paso 5: Filtrado de Datos**
Primero, identifica a las personas que tienen un puntaje alto en ansiedad (puedes elegir un valor alto). Luego, clasifica a las personas según las horas de sueño que reportan (menos de 5 horas, entre 5 y 7 horas, y más de 7 horas) y analiza cómo varía el nivel de estrés en cada grupo.

```python
#Personas que tienen un puntaje alto en ansiedad (arriba de 15, la escala presentada en el dataset va de 1-20)
high_anxiety = df_mh_alter[df_mh_alter['Anxiety_Score'] > 15]
high_anxiety
```

```python
no_ha=high_anxiety.shape[0]
perc_no_ha=(no_ha/df_mh_alter.shape[0])*100
print("Porcentaje de personas con puntaje alto en ansiedad: {0:.2f}%".format(perc_no_ha))
```

```python
#Personas que duermen menos de 5 horas
low_sleep = df_mh_alter[df_mh_alter['Sleep_Hours'] < 5]
low_sleep.sort_values(['Sleep_Hours'], ascending=True)
low_sleep[['Sleep_Hours','Stress_Level']]
```

```python
#Personas que duermen entre 5 y 7 horas
medium_sleep = df_mh_alter[(df_mh_alter['Sleep_Hours'] >= 5) & (df_mh_alter['Sleep_Hours'] <= 7)]
medium_sleep.sort_values(['Sleep_Hours'], ascending=True)
medium_sleep[['Sleep_Hours','Stress_Level']]
```

```python
#Personas que duermen más de 7 horas
high_sleep = df_mh_alter[df_mh_alter['Sleep_Hours'] > 7]
high_sleep.sort_values(['Sleep_Hours'], ascending=True)
high_sleep[['Sleep_Hours','Stress_Level']]
```
